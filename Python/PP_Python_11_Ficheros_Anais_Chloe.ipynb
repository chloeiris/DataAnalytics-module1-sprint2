```python
import os

def gestionar_carpetas(car1, car1_1, new_name_car1):
    ruta_inicial= os.getcwd()
    print(f'Tu directorio acutal es:{ruta_inicial}')
    
    if car1 not in os.listdir():
        os.mkdir(car1)
        print(f'Nueva carpeta {car1} creada')
    else:
        print(f'La carpeta {car1} ya existe')

    os.chdir(f'./{car1}')
    if car1_1 not in os.listdir():
        os.mkdir(car1_1)
        print(f'Nueva carpeta {car1_1} dentro de {car1} creada')
    else:
        print(f'La carpeta {car1_1} dentro de {car1} ya existe')

    os.chdir(f'./{car1_1}')  
    print(f'Tu directorio acutal es: {os.getcwd()}')
    
    os.chdir(ruta_inicial)
    os.rename(car1, new_name_car1)
    print(f'''Carpeta '{car1}' renombrada a '{new_name_car1}'.
            Tu directorio actual es: {os.getcwd()}''')
    
```

1. La funci贸n debe recibir 4 par谩metros:

    - C贸mo queremos leer el fichero.

2. Condiciones de la funci贸n:

 Pistas 
- Para chequear si un fichero existe, podemos usar la funci贸n os.listdir() que devuelve una lista con los nombres de los ficheros que hay en un directorio.
- Para crear un fichero, podemos usar la funci贸n open() que recibe como par谩metros el nombre del fichero, el modo de apertura y el encoding.
- Para chequear si el fichero existe deb茅is usar los for loops y el m茅todo in para comprobar si el nombre del fichero est谩 en la lista de ficheros.
- Usar un input para preguntar al usuario si quiere sobreescribir el fichero.
- Para sobreescribir el fichero, podemos usar la funci贸n open() que recibe como par谩metros el nombre del fichero, el modo de apertura y el encoding.
- Strings que cubren m谩s que una l铆nea tienen que ser iniciados y cerrados con tres comillas """.





```python
def escribir_fichero(nombre, contenido, mode_usr = 'r', encoding_usr = 'utf-8'):
    ubicacion = input('驴En qu茅 ubicaci贸n est谩 su archivo? Si no existe lo crearemos. Ej: ./primera-toma-contacto/datos/')
    try:
        if nombre not in os.listdir(ubicacion):
            # crear
            with open(ubicacion + nombre, 'x', encoding = encoding_usr) as f:
                #insertar contenido
                f.write(contenido)
                f.close()
                #mostrar
            with open(ubicacion + nombre, encoding = encoding_usr) as f:   # hay que volver a abrirlo, porque el modo anterior estaba en modo 'x' y ahora queremos leerlo
                leer1 = f.read()
                print(leer1)
        else:
            input_user = input('驴Desea sobreescribirlo? s/n')
            while input_user.lower() not in ['s','n']:                      # para asegurar que el usuario introduce la opci贸n adecuada
                print('Opci贸n no v谩lida, elige s/n')
                input_user = input('驴Desea sobreescribirlo? s/n')
            if input_user == 's':
                # sobreescribe el fichero
                with open(ubicacion + nombre, 'w', encoding = encoding_usr) as f:
                    # insertad contenido
                    f.write(contenido)
                    f.close()
                    # leedlo
                with open(ubicacion + nombre, encoding = encoding_usr) as f:
                    leer2 = f.read()
                    print(leer2)
    except:
        print('Directorio no encontrado.')
            


```
